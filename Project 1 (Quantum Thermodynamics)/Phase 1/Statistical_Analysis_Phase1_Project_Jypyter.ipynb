```python
################################## Feature Distribution
for i in range(1, 17):

    feature = df['fp hamiltonian'+str(i)].to_numpy()

    sns.displot(feature)

    # print(np.var(feature))
    plt.xlabel('Distribution of fp hamiltonian'+str(i))
    plt.ylabel('Value')

    plt.savefig('Distribution of fp hamiltonian' + str(i) + '.png')
```
